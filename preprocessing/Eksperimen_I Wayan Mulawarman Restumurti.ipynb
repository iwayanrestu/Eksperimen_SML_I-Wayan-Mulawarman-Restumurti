```python
#Drop ID Column
df_clean = df.drop(columns=['CUST_ID'])
```

```python
# handling missing values
df_clean.isnull().sum()
```

```python
df_clean['MINIMUM_PAYMENTS'].fillna(df_clean['MINIMUM_PAYMENTS'].median(), inplace=True)
df_clean['CREDIT_LIMIT'].fillna(df_clean['CREDIT_LIMIT'].median(), inplace=True)
```

```python
#Duplikat
df_clean.duplicated().sum()
```

```python
df_clean.skew().sort_values(ascending=False)
```

```python
skewed_cols = df_clean.skew()[df_clean.skew() > 1].index
df_log = df_clean.copy()
df_log[skewed_cols] = np.log1p(df_log[skewed_cols])
```

```python
#Scaling
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_log)

df_scaled = pd.DataFrame(df_scaled, columns=df_log.columns)
```
